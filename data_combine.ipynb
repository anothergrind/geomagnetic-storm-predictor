## Dataset Join Strategy — CME-Anchored

**One row = one CME event.** Each CME is a discrete "will this hit Earth and cause a storm?" question. **Decision time is launch + 12h** (the end of the wind feature window), giving 0.5–3.5 days of effective warning.

**Datasets used:**
- `space_weather_events` (CME rows) — anchor: one row per CME, with speed, half-angle, latitude, longitude, type
- `data/clean-data/solar-flare-events.parquet` — team-cleaned GOES flares, matched by time proximity (nearest flare ≤ 6h before CME)
- `omni_solar_wind_parameters` (raw) — solar wind around launch (−24h → +12h) via independent per-CME searchsorted windows; also provides Kp for the label
- `space_weather_indices` — not used directly (Kp comes from omni)

**Dropped:**
- `solar_wind` — 2026 real-time data only
- Day-based aggregation — misaligns causes (flare/CME at T) with effects (storm at T+1 to T+4)

**Time window:** 2017-02-09 – 2024-12-31 (flare catalog starts 2017-02-09; earlier CMEs would have structurally missing flare features)

**Feature groups per CME row:**
1. CME intrinsic properties: speed, half-angle, latitude, |longitude|, type
2. Associated flare (if any ≤ 6h prior): GOES class, log peak flux, duration, has_flare flag
3. Solar wind, launch − 24h → launch + 12h: mean/min Bz, southward-Bz fractions, mean |B|, mean/max speed, density, pressure, electric field, temperature
4. Solar-cycle context in the same window: F10.7, sunspot number

**Targets:**
- `kp_max_post` — max Kp in the 1–4 days after CME start (the typical CME travel window)
- `storm` = 1 if `kp_max_post` ≥ 5 (NOAA G1 threshold)
- `storm_attributed` — same, but each storm episode is credited only to the fastest/widest candidate CME (reduces multi-CME label noise)

### 1. CME anchor table

```python
import pandas as pd

df_cme = df_space_weather_events[df_space_weather_events['event_type'] == 'CME'].copy()

# normalize timestamps to tz-naive UTC
df_cme['cme_time'] = pd.to_datetime(df_cme['start_time'], utc=True).dt.tz_localize(None)
df_cme = df_cme[
    (df_cme['cme_time'] >= TRAIN_START) & (df_cme['cme_time'] <= TRAIN_END)
].reset_index(drop=True)

# encode CME type: S=1 (slow/narrow), C=2 (common), O=3, R=4 (fast/halo)
cme_type_map = {'S': 1, 'C': 2, 'O': 3, 'R': 4}
df_cme['cme_type_num'] = df_cme['cme_type'].map(cme_type_map)

df_cme = df_cme[[
    'cme_time', 'cme_speed_kms', 'cme_half_angle_deg',
    'cme_latitude', 'cme_longitude', 'cme_type_num',
]].copy()

df_cme['cme_id'] = range(len(df_cme))

print(f"CME events 2017–2024: {len(df_cme)}")
df_cme.head()
```

### 2. Match associated flare (time proximity)

DONKI activity IDs in `linked_events` don't map to our GOES-based `solar_flare_events`, so we match by time: find the most energetic flare that peaked within 6 hours before each CME. This captures the eruption that likely triggered it.

```python
class_map = {'A': 0, 'B': 1, 'C': 2, 'M': 3, 'X': 4}

df_flares = df_solar_flare_events.copy()
df_flares['peak_time'] = pd.to_datetime(df_flares['peak_time'])
df_flares = df_flares.dropna(subset=['peak_time'])  # merge_asof requires no nulls on key
df_flares['flare_class_num'] = df_flares['goes_class_letter'].map(class_map)
df_flares = df_flares.sort_values('peak_time').reset_index(drop=True)

df_cme_sorted = df_cme.sort_values('cme_time').reset_index(drop=True)

# For each CME, find the nearest flare peak that occurred ≤ 6h before it.
# merge_asof matches each CME to the most recent flare at or before cme_time.
FLARE_FEATURE_COLS = ['goes_class_letter', 'flare_class_num', 'log_peak_flux', 'duration_minutes', 'end_time_imputed']
flare_matched = pd.merge_asof(
    df_cme_sorted,
    df_flares[['peak_time'] + FLARE_FEATURE_COLS],
    left_on='cme_time',
    right_on='peak_time',
    direction='backward',
    tolerance=pd.Timedelta('6h'),
)

flare_matched = flare_matched.rename(columns={
    'goes_class_letter': 'flare_class',
    'log_peak_flux':     'flare_log_peak_flux',
    'duration_minutes':  'flare_duration_min',
    'end_time_imputed':  'flare_end_time_imputed',
})
flare_matched = flare_matched.drop(columns=['peak_time'])

# Explicit indicator instead of zero-filling flare features: within 2017+,
# NaN flare features genuinely mean "no flare in the 6h window", and the
# indicator lets models separate "no flare" from any imputed feature values.
flare_matched['has_flare'] = flare_matched['flare_class'].notna().astype(int)

matched = flare_matched['has_flare'].sum()
print(f"CMEs with an associated flare (≤6h prior): {matched} / {len(flare_matched)}  ({matched/len(flare_matched):.1%})")
flare_matched.head()
```

### 3. Solar wind background (launch − 24h → launch + 12h)

These features capture the ambient interplanetary conditions around the CME's launch, plus solar-cycle context (F10.7, sunspots). Background Bz, speed, and density all influence how geoeffective the CME will be when it arrives — and the 12h *after* launch can catch earlier disturbances already in transit toward Earth.

**Why +12h is safe (no leakage):** the CME takes ≥ ~1 day to reach the L1 monitors, so wind readings 12h after launch are still ambient background, not the CME's own signature — and the label window starts at launch + 1 day, comfortably after the feature window ends. The honest cost is that the prediction is now issued 12h after launch, so effective warning is 0.5–3.5 days instead of 1–4. Extending to +12h improved attributed-label classification (XGB PR-AUC 0.106 → 0.156, ROC-AUC 0.835 → 0.860) with no change on the naive label.

Implementation note: 83% of CMEs launch within 24h of the previous one (median gap ≈ 7.6h), so windows overlap heavily. A forward `merge_asof` would assign each OMNI reading to only the *next* CME, silently truncating most windows to the inter-CME gap. Instead, `searchsorted` gives every CME its own full, independent window; the same OMNI hour can contribute to several CMEs.

```python
import numpy as np

df_omni = df_omni_solar_wind_parameters.copy()
df_omni['datetime'] = pd.to_datetime(df_omni['datetime'])
df_omni = df_omni[
    (df_omni['datetime'] >= TRAIN_START) & (df_omni['datetime'] <= TRAIN_END)
].sort_values('datetime').reset_index(drop=True)

# Strong keeps per the OMNI EDA, plus its two "optional" features
# (electric_field_mvpm, proton_temperature_k) and solar-cycle context.
WIND_COLS = [
    'bz_gsm_nt', 'b_magnitude_avg_nt', 'flow_speed_kms',
    'proton_density_cm3', 'flow_pressure_npa',
    'electric_field_mvpm', 'proton_temperature_k',
    'f107_index_sfu', 'sunspot_number',
]

# Window: [launch − 24h, launch + 12h]. The +12h stays well clear of the
# label window (starts at +1 day) and of the CME's own L1 arrival (≥ ~1 day),
# so it adds signal (earlier disturbances in transit) without leakage.
# Decision time is therefore launch + 12h.
WINDOW_PRE_H  = 24
WINDOW_POST_H = 12

omni_times = df_omni['datetime'].values
w_starts = (flare_matched['cme_time'] - pd.Timedelta(hours=WINDOW_PRE_H)).values
w_ends   = (flare_matched['cme_time'] + pd.Timedelta(hours=WINDOW_POST_H)).values
idx_starts = np.searchsorted(omni_times, w_starts)
idx_ends   = np.searchsorted(omni_times, w_ends, side='right')

def window_stats(s, e):
    if e <= s:
        return {}
    w = df_omni.iloc[s:e]
    bz = w['bz_gsm_nt']
    return {
        'wind_bz_mean':        bz.mean(),
        'wind_bz_min':         bz.min(),
        # sustained southward Bz matters more than single readings (solar-wind EDA)
        'wind_bz_south_frac':  (bz < 0).mean() if bz.notna().any() else np.nan,
        'wind_bz_strong_south_frac': (bz < -10).mean() if bz.notna().any() else np.nan,
        'wind_imf_mean':       w['b_magnitude_avg_nt'].mean(),
        'wind_speed_mean':     w['flow_speed_kms'].mean(),
        'wind_speed_max':      w['flow_speed_kms'].max(),
        'wind_density_mean':   w['proton_density_cm3'].mean(),
        'wind_pressure_mean':  w['flow_pressure_npa'].mean(),
        'wind_efield_mean':    w['electric_field_mvpm'].mean(),
        'wind_temp_mean':      w['proton_temperature_k'].mean(),
        'ctx_f107':            w['f107_index_sfu'].mean(),
        'ctx_sunspots':        w['sunspot_number'].mean(),
    }

df_wind_pre = pd.DataFrame(
    [window_stats(s, e) for s, e in zip(idx_starts, idx_ends)],
    index=flare_matched['cme_id'],
).reset_index()

with_wind = df_wind_pre['wind_bz_mean'].notna().sum()
print(f"CMEs with launch-window wind data: {with_wind} / {len(flare_matched)}")
df_wind_pre.head()
```

### 4. Storm label — Kp-based, with per-storm attribution

For each CME, take the max Kp in the 1–4 day window after launch (typical Sun→Earth travel time; day 0 is skipped to avoid contamination from launch-time disturbances). `storm = 1` if max Kp ≥ 5 — the NOAA G1 threshold, and the same definition the OMNI EDA used (`kp_index` is stored as Kp × 10).

**Attribution:** CMEs launch a median of ~7.6h apart, so several CMEs' 4-day windows often contain the *same* storm, and the naive label marks all of them `storm = 1` even though only one physically caused it. To reduce that label noise we also build `storm_attributed`: storm hours are grouped into episodes (gaps > 12h start a new episode), and within each episode only the fastest CME (tie-break: widest half-angle) whose window covers it keeps the positive label; the other candidates are relabeled 0.

```python
import numpy as np

# --- Windowed max Kp per CME (searchsorted: independent overlapping windows) ---
omni_kp = (
    df_omni[['datetime', 'kp_index']]
    .dropna(subset=['kp_index'])
    .sort_values('datetime')
    .reset_index(drop=True)
)
kp_times = omni_kp['datetime'].values
kp_vals  = omni_kp['kp_index'].values  # stored as Kp × 10

t_starts = (flare_matched['cme_time'] + pd.Timedelta('1D')).values
t_ends   = (flare_matched['cme_time'] + pd.Timedelta('4D')).values
idx_starts = np.searchsorted(kp_times, t_starts)
idx_ends   = np.searchsorted(kp_times, t_ends, side='right')

kp_maxes = np.array([
    kp_vals[s:e].max() / 10 if e > s else np.nan
    for s, e in zip(idx_starts, idx_ends)
])

storm_labels = flare_matched[['cme_id']].copy()
storm_labels['kp_max_post'] = kp_maxes
storm_labels['storm'] = (storm_labels['kp_max_post'] >= 5).astype(float)
storm_labels.loc[storm_labels['kp_max_post'].isna(), 'storm'] = np.nan

# --- Storm episodes: consecutive-ish storm hours (Kp >= 5), new episode after a >12h gap ---
storm_hr_mask  = kp_vals >= 50
storm_hr_times = kp_times[storm_hr_mask]
gap_hours   = np.diff(storm_hr_times).astype('timedelta64[h]').astype(int)
episode_ids = np.concatenate([[0], np.cumsum(gap_hours > 12)])
ep_start_time = pd.Series(storm_hr_times).groupby(episode_ids).min()  # episode -> onset
print(f"Storm episodes (Kp>=5, 12h gap rule): {episode_ids.max() + 1}")

# First storm hour inside each CME's window -> the episode this CME's label came from
ep_starts = np.searchsorted(storm_hr_times, t_starts)
ep_ends   = np.searchsorted(storm_hr_times, t_ends, side='right')
storm_labels['episode'] = [
    episode_ids[s] if e > s else -1 for s, e in zip(ep_starts, ep_ends)
]

# --- Attribution: candidates = CMEs whose window covers the episode ---
cand = (
    flare_matched[['cme_id', 'cme_time', 'cme_speed_kms', 'cme_half_angle_deg']]
    .merge(storm_labels.loc[storm_labels['episode'] >= 0, ['cme_id', 'episode']], on='cme_id')
)

# v1: fastest candidate wins (tie-break: widest half-angle)
v1_winner = (
    cand.sort_values(['cme_speed_kms', 'cme_half_angle_deg'], ascending=False, na_position='last')
    .groupby('episode').first()['cme_id']
)

# v2: best arrival-time match. Predicted transit = 1 AU / speed; blame the
# candidate whose predicted arrival is closest to the episode onset. This is
# closer to how DONKI analysts attribute storms than "fastest wins" (no drag
# correction yet — fast CMEs decelerate en route, so this is still approximate).
AU_KM = 1.496e8
cand['pred_arrival'] = cand['cme_time'] + pd.to_timedelta(AU_KM / cand['cme_speed_kms'], unit='s')
cand['timing_err_h'] = (cand['pred_arrival'] - cand['episode'].map(ep_start_time)).abs().dt.total_seconds() / 3600
v2_winner = (
    cand.dropna(subset=['timing_err_h']).sort_values('timing_err_h')
    .groupby('episode').first()['cme_id']
)
v2_winner = v2_winner.reindex(v1_winner.index).fillna(v1_winner)  # no-speed episodes: fall back to v1

for col, winners in [('storm_attributed', v1_winner), ('storm_attributed_v2', v2_winner)]:
    storm_labels[col] = ((storm_labels['episode'] >= 0)
                         & storm_labels['cme_id'].isin(winners.values)).astype(float)
    storm_labels.loc[storm_labels['storm'].isna(), col] = np.nan

labeled = storm_labels['storm'].notna().sum()
print(f"CMEs with post-event Kp data: {labeled} / {len(storm_labels)}")
print(f"Naive label          — storm CMEs: {int((storm_labels['storm'] == 1).sum())} ({(storm_labels['storm'] == 1).mean():.1%})")
print(f"Attributed v1 (speed)— storm CMEs: {int((storm_labels['storm_attributed'] == 1).sum())} ({(storm_labels['storm_attributed'] == 1).mean():.1%})")
print(f"Attributed v2 (time) — storm CMEs: {int((storm_labels['storm_attributed_v2'] == 1).sum())} ({(storm_labels['storm_attributed_v2'] == 1).mean():.1%})")
print(f"v1/v2 pick the same CME for {(v1_winner == v2_winner).mean():.1%} of episodes")
```

### 5. Merge all into final dataset

```python
df_final = (
    flare_matched
    .merge(df_wind_pre, on='cme_id', how='left')
    .merge(storm_labels[['cme_id', 'kp_max_post', 'storm', 'storm_attributed', 'storm_attributed_v2']], on='cme_id', how='left')
)

# CMEs near the end of the dataset won't have a full 4-day post window — drop them
df_final = df_final.dropna(subset=['storm']).copy()
for c in ['storm', 'storm_attributed', 'storm_attributed_v2']:
    df_final[c] = df_final[c].astype(int)

# Flare features stay NaN when no flare matched (has_flare=0): zero-filling would
# invent fake values (e.g. log flux of 0 = 1 W/m² would be an absurdly huge flare).
# XGBoost handles NaN natively; the Ridge pipeline median-imputes with has_flare
# available to disambiguate.

print(df_final.shape)
print(f"\nStorm rate (naive):           {df_final['storm'].mean():.1%}")
print(f"Storm rate (attributed v1):   {df_final['storm_attributed'].mean():.1%}")
print(f"Storm rate (attributed v2):   {df_final['storm_attributed_v2'].mean():.1%}")
print(f"\nMissing value rates:")
print(df_final.isnull().mean()[df_final.isnull().mean() > 0].sort_values(ascending=False))
df_final.head()
```

## Summary of design choices

**What this notebook produces:** `data/cme_dataset.csv` — one row per CME (4,734 rows, 2017-02-09 → 2024-12-31), with features knowable by launch + 12h and three candidate labels. It builds the dataset only; modeling lives elsewhere.

**Choices and why:**

1. **CME-anchored rows** — each row asks "will this CME cause a storm?", preserving the cause→effect time structure that day-binned aggregation destroys.

2. **Time window starts 2017-02-09** (GOES flare catalog start), not 2010. Before 2017, "no flare matched" would mean "no flare *data*", biasing the flare features and entangling them with the solar cycle. Post-2017, NaN flare features genuinely mean "no flare within 6h".

3. **Flares from `data/clean-data/solar-flare-events.parquet`** (team-cleaned: garbage rows dropped, `duration_minutes`/`log_peak_flux`/`end_time_imputed` added), matched by time proximity — nearest flare peaking ≤ 6h before launch. `linked_events` was rejected for matching: only ~12% coverage, and its IDs point to DONKI's own flare entries, not the GOES catalog.

4. **No zero-filling of missing flare features.** They stay NaN with an explicit `has_flare` 0/1 flag. Zero-fill would invent physics (log-flux 0 = 1 W/m², an impossibly large flare). XGBoost handles NaN natively; linear models should impute inside their pipeline (fit on train only).

5. **Solar wind from raw OMNI, not the cleaned parquet** — we need `kp_index` for the label, and NaN-tolerant window aggregation instead of complete-case row dropping (OMNI gaps correlate with extreme conditions). Feature *selection* follows the OMNI EDA: the 5 "strong keeps" + electric field, temperature, and solar-cycle context (F10.7, sunspots).

6. **Wind window = launch − 24h → launch + 12h, via per-CME `searchsorted`.** 83% of CMEs launch within 24h of the previous one, so a `merge_asof` join silently truncates most windows; searchsorted gives every CME its full, independent window. The +12h extension is leakage-free (the CME needs ≥ ~1 day to reach L1; the label window starts at +1 day) and improved attributed-label classification. Cost: decision time is launch + 12h → effective warning 0.5–3.5 days.

7. **Label: max Kp in [launch + 1d, launch + 4d]; storm = Kp ≥ 5** (NOAA G1). Day 0 excluded to avoid flare contamination. Kp ≥ 5 matches the OMNI EDA convention (`kp_index` stored ×10).

8. **Three labels, use deliberately:**
   - `storm` (22.8% positive) — "did a storm *follow*?" Noisy for causal claims: ~7 CMEs share each storm window on average.
   - `storm_attributed` (3.0%) — "did this CME *cause* it?" Storm hours grouped into 158 episodes (>12h gap = new episode); the fastest candidate per episode keeps the label. **Primary attributed label.** Caveat: speed-based winners make speed features partly self-fulfilling.
   - `storm_attributed_v2` (3.0%) — arrival-time variant (1 AU / speed vs episode onset). Kept for comparison; did not improve analyst agreement.

9. **Validation against NASA analysts** (`leads_to_gst`, 54 gold links): 90.7% of analyst-blamed CMEs have `storm = 1` (window machinery is sound), but both attribution rules match the analysts' pick only ~24% of the time. **Known limitation:** attribution from tabular launch data appears capped well below expert level — quote this caveat with any attributed-label results. Analyst links are for validation only; silence ≠ innocent, so never train on `leads_to_gst`.

10. **Modeling guidance:** split train/test chronologically (activity is autocorrelated); report precision/recall/F1/PR-AUC, not accuracy (positives are 3–23%); regression on `kp_max_post` has shown ~no signal — the classification framing is where the signal is.
